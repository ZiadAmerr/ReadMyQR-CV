```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image

def calculate_average(pixels, y_tuple, x_tuple):
    count = 0
    sum = 0
    for i in range(y_tuple[0], y_tuple[1]):
        for j in range(x_tuple[0], x_tuple[1]):
            sum += pixels[i][j]
            count += 1
    print(count)
    
    return sum / count

def detect_inversion(image):

    X_TOP_LEFT = (1, (7 * 48) - 1)
    Y_TOP_LEFT = ((7 * 48) + 2, (8 * 48) - 2)

    # Convert the image into a 2D numpy array
    pixels = np.array(image.getdata()).reshape((image.size[1], image.size[0]))
    
    # Calculate the average intensity of the 8th row from above in cells

    average = calculate_average(pixels, Y_TOP_LEFT, X_TOP_LEFT)
    print(average)
    # Determine inversion based on average intensity
    if int(average) > 230:
        return "NI" #NI = Not Inverted
    elif int(average) < 30:
        return "I" #I = Inverted
    else:
        return "N/A" 

def detect_rotation(image):

    X_TOP_LEFT = (1, (7 * 48) - 1)
    Y_TOP_LEFT = ((7 * 48) + 2, (8 * 48) - 2)

    X_TOP_RIGHT = (1007 - (7 * 48) - 1, 1007 - 1)
    Y_TOP_RIGHT = ((7 * 48) + 2, (8 * 48) - 2)

    X_BOTTOM_LEFT = (1, (7 * 48) - 1)
    Y_BOTTOM_LEFT = (1007 - ((8 * 48) - 2), 1007 - ((7 * 48) + 2))


    
    # Convert the image into a 2D numpy array
    pixels = np.array(image.getdata()).reshape((image.size[1], image.size[0]))
    
    # Calculate the average intensity of the 8th row from above in cells

    average_top_left = calculate_average(pixels, Y_TOP_LEFT, X_TOP_LEFT)
    average_top_right = calculate_average(pixels, Y_TOP_RIGHT, X_TOP_RIGHT)
    average_bottom_left = calculate_average(pixels, Y_BOTTOM_LEFT, X_BOTTOM_LEFT)
    
    print(average_top_left)
    print(average_top_right)
    print(average_bottom_left)

    average_total = (average_top_left + average_top_right + average_bottom_left) / 3

    # Determine inversion based on average intensity
    if int(average_total) > 230:
        return "NR" #NR = Not Rotated
    elif int(average_total) < 30:
        return "R" #R = Rotated
    else:
        return "N/A"
```

```python
from PIL import Image

# Open the image
filepath = "resized_01-Getting-started.png.png"

# image = cv2.imread(
#     f"resized/{filepath}", cv2.IMREAD_GRAYSCALE)
# plt.imshow(image, cmap='gray')

image = Image.open(f"resized/{filepath}")

# Define the coordinates for the center of the red dot
center_x = 1007 - 1  # Example coordinate
center_y = int(7 * 48) + 2  # Example coordinate
dot_size = 1  # Example dot size


"""
FOR TOP LEFT:
x = 1 ---> (7 * 48) - 1
y = (7 * 48) + 2 ---> (8 * 48) - 2

FOR TOP RIGHT
x = 1007 - 1 <--- 1007 - (7 * 48) - 1
y = (7 * 48) + 2 ---> (8 * 48) - 2

FOR BOTTOM LEFT
x = 1 ---> (7 * 48) - 1
y = 1007 - [(7 * 48) + 2] ---> 1007 - [(8 * 48) - 2]

AREA: TOP LEFT: 
"""




# Define the RGB color for red
red = (128)

# Iterate over the pixels in the dot region and set them to red
for i in range(center_x - dot_size // 2, center_x + dot_size // 2 + 1):
    for j in range(center_y - dot_size // 2, center_y + dot_size // 2 + 1):
        image.putpixel((i, j), red)

# Show or save the modified image
image.show()
# image.save("path/to/your/output/image_with_dot.jpg")
```
