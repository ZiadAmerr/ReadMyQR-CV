```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the blurry QR code image
blurry_qr_code = cv2.imread('test_data/9_blurred.png', cv2.IMREAD_GRAYSCALE)

# Apply thresholding to create a binary image
_, binary_image = cv2.threshold(blurry_qr_code, 127, 255, cv2.THRESH_BINARY)

# Apply morphological closing to sharpen the edges
kernel_size = 3  # Adjust kernel size as needed
kernel = np.ones((kernel_size, kernel_size), np.uint8)
closed_image = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel)

# Display the original blurry QR code, binary image, and closed image
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(blurry_qr_code, cmap='gray')
plt.title('Blurry QR Code')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image (Thresholding)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(closed_image, cmap='gray')
plt.title('Closed Image (Morphological Closing)')
plt.axis('off')

plt.show()
```
