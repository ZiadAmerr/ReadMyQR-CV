```python
from google.colab.patches import cv2_imshow
import cv2

# Read the image
image = cv2.imread("13-2el-noor-2ata3.png")

# Convert RGBA to Grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGBA2GRAY)

# Increase brightness by adding 150 to each pixel value
increased_brightness = cv2.add(gray_image, 150)

# Display the original and increased brightness images
cv2_imshow(increased_brightness)
```

```python
# Apply Histogram Equalization
equalized_image = cv2.equalizeHist(increased_brightness)
cv2_imshow(equalized_image)
```

```python
_, thresholded_image = cv2.threshold(equalized_image, 100, 255, cv2.THRESH_BINARY)


cv2_imshow(thresholded_image)
```

```python
kernel = cv2.getStructuringElement(
    cv2.MORPH_RECT, (43, 43)
)  # You can adjust the kernel size as needed

# Apply closing
eroded_image = cv2.erode(equalized_image, kernel)
cv2_imshow(eroded_image)
```
