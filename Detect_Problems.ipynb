```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
def plot_histogram(image):
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])
    plt.plot(hist)
    plt.title("Histogram")
    plt.xlabel("Pixel value")
    plt.ylabel("Frequency")
    plt.show()

def unique_pixel_values(image):
    # Flatten the image to a 1D array
    flattened_image = image.ravel()

    # Find unique pixel values and their counts
    unique_values, counts = np.unique(flattened_image, return_counts=True)

    # Create a dictionary to store unique pixel values and their counts
    pixel_counts = dict(zip(unique_values, counts))

    return pixel_counts
```

```python
# Define the path to the folder containing the images
folder_path = "test_cases"

# Initialize an empty dictionary to store the images
image_dict = {}
target_size = (1012, 1012)
# Iterate over each file in the folder
for filename in os.listdir(folder_path):
      # Open the image using cv2
      image_path = os.path.join(folder_path, filename)
      image = cv2.imread(image_path,  cv2.IMREAD_GRAYSCALE)
      image = cv2.resize(image, target_size)

      # Add the image to the dictionary with the filename as the key
      image_dict[filename] = (image, unique_pixel_values(image))

# Sort the dictionary keys (filenames)
sorted_keys = sorted(image_dict.keys())

# Create a new dictionary with sorted keys
sorted_image_dict = {key: image_dict[key] for key in sorted_keys}

# Iterate over the sorted dictionary and print each key-value pair
for filename, pair in sorted_image_dict.items():
    print(f"Filename: {filename}")
    plt.figure()
    plt.imshow(pair[0], cmap = 'gray')
    plt.title(filename)
    plt.axis('off')
    plt.show()
    print("-----------------------------")
```

```python
# Plot histogram for each image
for name, pair in sorted_image_dict.items():
    print(f"{name}:")
    plot_histogram(pair[0])
    print(pair[1])
```

```python
def filter_images_by_metric(image_dict):
    filtered_dict = {}

    for image_name, (image, pixel_counts) in image_dict.items():
        total_pixels = sum(pixel_counts.values())
        zero_pixels = pixel_counts.get(0, 0)
        white_pixels = pixel_counts.get(255, 0)
        percentage_zero_or_white = (zero_pixels + white_pixels) / total_pixels * 100

        if percentage_zero_or_white >= 92:
            filtered_dict[image_name] = (image, pixel_counts)

    return filtered_dict

# Example usage:
filtered_images = filter_images_by_metric(sorted_image_dict)

for key in filtered_images.keys():
    if key in sorted_image_dict and sorted_image_dict[key] == filtered_images[key]:
        del sorted_image_dict[key]
```

```python
nominator = 0
denominator = 0
means = []
for name, (_, pixel_values) in sorted_image_dict.items():
  for value, number_of_pixels in pixel_values.items():
    nominator += value * number_of_pixels
    denominator += number_of_pixels
  mean = nominator/denominator
  print(f"{name}: {mean}")
  nominator = 0
  denominator = 0
  means.append((name, mean))
```

```python
hist_equlize = {}
for (name, mean) in means:
  if mean < 35 or mean > 195:
    if name in sorted_image_dict:
        # Add the key-value pair to the new dictionary
        hist_equlize[name] = sorted_image_dict[name]
        # Remove the key-value pair from the original dictionary
        del sorted_image_dict[name]

print("sorted_image_dict Dictionary after removal:", sorted_image_dict.keys())
print("hist_equlize Dictionary:", hist_equlize.keys())
```

```python
def detect_salt_and_pepper(image_path, filter_size=3, threshold=1000):
    # Load the image
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Calculate variance of the original image
    variance_original = np.var(img)

    # Apply median filtering to smooth the image
    median_img = cv2.medianBlur(img, filter_size)

    # Calculate variance of the filtered image
    variance_filtered = np.var(median_img)

    
    # Calculate the absolute difference in variance
    diff_variance = np.abs(variance_original - variance_filtered)
    print(f"{image[0]} variance: {diff_variance}")
    # Determine if the image likely contains salt and pepper noise based on the difference in variance
    if diff_variance > threshold:
        return True  
    else:
        return False  
salt_pepper_dict = {}

# Iterate over a copy of sorted_image_dict to avoid modification during iteration
for name, (image, pixel_values) in sorted_image_dict.items():
    image_path = f'./test_cases/{name}'
    has_salt_and_pepper = detect_salt_and_pepper(image_path)
    
    if has_salt_and_pepper:
        salt_pepper_dict[name] = (image, pixel_values)
for name, (image, pixel_values) in salt_pepper_dict.items():
    if name in sorted_image_dict:
        del sorted_image_dict[name]


```
