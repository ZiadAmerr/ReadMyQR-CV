```python
import cv2
import numpy as np

# Read the image
image = cv2.imread('12-mal7-w-felfel.png')

# Apply median filter with kernel size 5x5

# Apply Gaussian blur with kernel size 9x9 and sigmaX 0 (automatic calculation)
gaussian_filtered = cv2.GaussianBlur(image, (45, 45), 0)

# Convert the Gaussian blurred image to grayscale
gray_image = cv2.cvtColor(gaussian_filtered, cv2.COLOR_BGR2GRAY)

# Apply thresholding with a threshold value of 127
ret, thresholded_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

# Define a kernel for morphological operations
kernel = np.ones((5, 5), np.uint8)


median_filtered_5 = cv2.medianBlur(thresholded_image, 45)

# Apply closing operation
opeining = cv2.morphologyEx(median_filtered_5, cv2.MORPH_OPEN, kernel)

# Apply opening operation MORPH_OPEN
closing = cv2.morphologyEx(thresholded_image, cv2.MORPH_CLOSE, kernel)

```

```python
# Display the thresholded image, closing, and opening results
cv2_imshow(thresholded_image)

```
