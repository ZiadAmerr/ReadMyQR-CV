```python
import cv2
from google.colab.patches import cv2_imshow

# Load an image from file in grayscale
image = cv2.imread('04-Black-mirror.png', 0)

# Check if image is loaded successfully
if image is not None:
    # Invert the colors
    inverted_image = 255 - image

    # Display the inverted grayscale image
    cv2_imshow(inverted_image)

else:
    print("Error: Unable to load image.")
```

```python
_, thresholded_image = cv2.threshold(inverted_image, 200, 255, cv2.THRESH_BINARY)
cv2_imshow(thresholded_image)
```

either -90 degrees rotation or vertical mirroring.

```python
rotated_image = cv2.rotate(thresholded_image, cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2_imshow(rotated_image)
```
