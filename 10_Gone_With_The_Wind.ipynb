```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

# Read the image
image = cv2.imread('10-Gone-With-The-Wind.png', cv2.IMREAD_GRAYSCALE)

# Subtract 100 from each pixel value
subtracted_image = np.clip(image - 170, 0, 255).astype(np.uint8)

# Apply Adaptive Threshold
# Display the original and thresholded images
cv2_imshow(subtracted_image)
```

```python
equalized_image = cv2.equalizeHist(subtracted_image)


cv2_imshow(equalized_image)
```

```python
_, thresholded_image = cv2.threshold(equalized_image, 127, 255, cv2.THRESH_BINARY)


cv2_imshow(thresholded_image)
```
